# **KNN on IRIS Dataset** using CSV file

```python
import pandas as pd
df=pd.read_csv('/content/Iris.csv')
print(df)
```

```python
X=df.drop('Species',axis='columns')
y=df['Species']
print(X)
print('------------------------------------------------')
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=9)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#if k =3,then 3 nearest neighbours.                                   #value of k is always odd.{if given even,then add 1 and solve the problem.}
model=KNeighborsClassifier(n_neighbors=3)         #value of k is given using parameter n_neighbour
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
print(model.score(X_test,y_test))
```

# **KNN on IRIS Dataset** using sklearn dataset

```python
import pandas as pd
from sklearn.datasets import load_iris
d=load_iris()             #formation of dictionary
#converting dictionary to dataframe.
df=pd.DataFrame(d['data'],columns=d['feature_names'])        #Since we are taking 'data' we only get independent data,thus Species(dependent value is missing.)
df['target']=d['target']                #Adding target in df. 
print(df)
```

```python
X=df.drop('target',axis='columns')
y=df['target']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=9)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,y_train)
model.predict(X_test)
print(model.score(X_test,y_test))
```

# **KNN on Digits Dataset(MNIST)** using CSV file




```python
#Digits Dataset. 
#If there is picture of pixles 480*600 then each pixel has its own intensity values,and intensity value ranges btween 0-255. 
df1=pd.read_csv('/content/mnist_test.csv',on_bad_lines='skip')              #to remove missing data we use on_bad_lines ='skip' or use drop method.
#without on_bad_lines it gives error due to missing entries.
df1.dropna(inplace=True)        #dropna removes missing entries.
```

```python
X=df1.drop('label',axis='columns') #removing dependent label variable column.
y=df1['label']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=9)
```

```python
model=KNeighborsClassifier(n_neighbors=10)
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
print(model.score(X_test,y_test))
```
