# **# Constructing Decision Tree of Dataframe "df2" with {Criterion='gini'} -->**




```python
#Importing digit dataset -->

import pandas as pd
from sklearn.datasets import load_digits
d = load_digits()
df2=pd.DataFrame(d['data'],columns=d['feature_names'])
df2['target']=d['target']
```

```python
# Intialising values of X and y -->

X=df2.drop('target',axis='columns')
y=df2['target']
```

```python
#Applying Training algorithm  -->

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
#Applying Decision Tree Algorithm --> {Creating model with criterion = 'entropy'}

from sklearn.tree import DecisionTreeClassifier 
model=DecisionTreeClassifier(criterion='gini')  
```

```python
#Training,Testing and finding Accuracy -->

model.fit(X_train,y_train)
model.predict(X_test)
print('Accuracy',model.score(X_train,y_train))
```

```python
#Constructing Decision Tree of our model with Matplotlib -->

import matplotlib.pyplot as plt
plt.figure(figsize=(25,20))    
from sklearn.tree import plot_tree
plot_tree(model,filled=True)
```
