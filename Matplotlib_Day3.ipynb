# **Scatter**

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Scatter-->gives scattered points in a graph.It only draws points not line.   
#In scatter we can give color with keyword 'c' and marker with keyword 'marker' .
#Type of line and marker size cannot be given in scatter.

x=[10,20,30,40,50]
y=[1,2,3,4,5]
plt.scatter(x,y)
```

```python
df=pd.read_csv('/content/IRIS.csv')

plt.scatter(df.sepal_length,df.sepal_width)
plt.show()
```

```python
#Data of two differnet datasets can also be represented in single graph.
x=[10,20,30,40,50]
y=[1,2,3,4,5]
plt.scatter(x,y,color='red')
p=[1,1,1,1,1]
q=[1,2,3,4,5]
plt.scatter(p,q,color='green')
plt.show()
```

```python
# We can assign differmet color for each point with 'color' keyword.
x=[10,20,30,40,50]
y=[1,2,3,4,5]
plt.scatter(x,y,color=['red','green','blue','pink','cyan'])  
plt.show()
```

```python
import random              # used to random outcome after each execution.
x=[10,20,30,40,50]
y=[1,2,3,4,5]
plt.scatter(x,y,c=[random.randint(0,100) for i in range(5)],cmap='viridis')
#Cmap{colormap} provides many built in colormaps,means it maps numbers to color. After every execution color of points in graph changes. 
#viridis is the default colormap. 
plt.show()
```

```python
x=[10,20,30,40,50]
y=[1,2,3,4,5]
plt.scatter(x,y,s=[10,20,40,80,160])          # 's' stands for marker size.
plt.show()
```

```python
#Alpha tells about transparency of a point. It's value is between 0 to 1.iIts type is float. For 0 point is transparent , with increase it becomes visible accordingly.
x=[10,20,30,40,50]
y=[1,2,3,4,5]
plt.scatter(x,y,s=[10,20,40,80,160],alpha=0.9)
plt.show()
```

# **Bars**

```python
categories=['A','B','C','D']
y=[10,20,30,40]
plt.bar(categories,y)        #bar chart --> along x axis categories are defined.
plt.show()
```

```python
#barh makes y axis x.  
categories=['A','B','C','D']
y=[10,20,30,40]
plt.barh(categories,y)
plt.show()
```

```python
#for diffeent color of each bar complete name is to be written.
categories=['A','B','C','D']
y=[10,20,30,40]
plt.bar(categories,y,color=['cyan','magenta','red','pink'])
plt.show()
```

```python
categories=['A','B','C','D']
y=[10,20,30,40]
plt.bar(categories,y,width=0.5)     #width of bars is given by attribute 'width'.
plt.show()
```

```python
categories=['A','B','C','D']
y=[10,20,30,40]
plt.barh(categories,y,height=.1)             #with barh bar heigth is assigned instead of bar width.
plt.show()
```

# **Piechart**

```python
a=[10,20,30,40]
b=['I','II','III','IV']
c=['red','green','blue','cyan']
plt.pie(a,labels=b,colors=c,startangle=0,explode=[0.2,0,0,0],shadow=True)   #shadow is a type of boolean value. USe to give shadow in piechart pies.
plt.show()
```

```python
a=[10,20,30,40]
b=['I','II','III','IV']
c=['red','green','blue','cyan']
plt.pie(a,labels=b,colors=c,startangle=0,explode=[0.2,0,0,0],shadow=True)    
plt.legend()         #legend is used to show details with categorues and color.
plt.show()
```

```python
a=[10,20,30,40]
b=['I','II','III','IV']
c=['red','green','blue','cyan']
plt.pie(a,labels=b,colors=c,startangle=0,explode=[0.2,0,0,0],shadow=True)
plt.legend(title='4 Categories:-')             #to give title to legend dialog box.
plt.show()
```

```python
a=[1,2,3,40]
b=['I','II','III','IV']
c=['red','green','blue','cyan']

#autopct shows percentage ,given with format specifier like %d %f,we can also give precision value in float like %0.1f. 
#format specifier ke end me do baar perecent sign aayega to show percent sign in piechart values ,without it numbers are displayed as integer instead of percentage.

plt.pie(a,labels=b,colors=c,startangle=0,autopct='%.1f%%')
plt.show()
```

```python
a=[1,2,3,40]
b=['I','II','III','IV']
c=['red','green','blue','cyan']

#wedgeprops is use to inseert border to piechart. 
#values are given through dictionaries. 
# Synatx --> wedgeeprops={'edgecolor':'color','linewidth':5}


plt.pie(a,labels=b,colors=c,startangle=0,autopct='%.1f%%',wedgeprops={'edgecolor':'magenta','linewidth':4})
plt.show()
```
