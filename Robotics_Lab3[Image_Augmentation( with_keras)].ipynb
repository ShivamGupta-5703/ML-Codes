```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import  array_to_img, img_to_array, load_img
```

```python
datagen = ImageDataGenerator(
        rotation_range = 40,
        shear_range = 0.2,
        zoom_range = 0.2,
        horizontal_flip = True,
        brightness_range = (0.5, 1.5))
```

```python
img = load_img('/content/4834078.jpeg')
```

```python
# Converting the input sample image to an array
x = img_to_array(img)
# Reshaping the input image
x = x.reshape((1, ) + x.shape) 
```

```python
i = 0
for batch in datagen.flow(x, batch_size = 1,                           #batch sie 1 means it takes one image at a time.
                          save_to_dir ='preview',                      # create a folder named preview , Images produced will be stored here.
                          save_prefix ='image', save_format ='jpeg'):
    i += 1
    if i > 5:
        break
```
