 #1. Importing libraries and loading dataset-->

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
companies = pd.read_csv("/content/Salary_Data.csv")
companies
```

 #2. Exploring dataset -->

```python
companies.describe()               #gives statistics of data
```

 #3. Finding correlations and printing them using Seaborn heatmap -->

```python
companies[["R&D Spend","Administration","Marketing Speed"]].corr()
sns.heatmap(companies[["R&D Spend","Administration","Marketing Speed"]].corr())
```

 #4. Categorising independent and dependent variables -->

```python
X = companies.iloc[:,:-1].values
y = companies.iloc[:,4].values
```

 #5. Removing highly correlated column -->

```python
companies.rop(["Marketing Speed"],axis=1,inplace=True)
companies
```

 #6. Encoding (Dummy encoding) -->

```python
dummy = pd.get_dummies(companies.state)
companies = pd.concat([companies,dummy],axis=1)
companies
```

```python
# dropping column state
companies.drop(['state'],axis=1,inplace=True)
companies
```

 #7. Feature Scaling -->

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
companies[['R&D Speed','Adminsitration']] = scale.fit_transform(companies[['R&D Speed','Administration']])
companies
```

 #8. Specifying X (Independent) and Y (Dependent) -->

```python
Y = companies.iloc[:,2]
print(Y)
print('---------------------------------------------------------')

companies.drop(['Profit'],axis=1,inplace=True)
print(companies)
print('---------------------------------------------------------')

X = companies.iloc[:,:]
print(X)
print('---------------------------------------------------------')

X = companies.iloc[:,:].values
print(X)
```

 #9. Spiliting into Training and Testing data -->

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
print(X_train)
print("-------------------------------------------------------------------")
print(X_test)
print("-------------------------------------------------------------------")
print(Y_train)
print("-------------------------------------------------------------------")
print(Y_test)
```

 #10. Applying Linear Regression Model -->[link text](https://)

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
X1=X_train.reshape(-1,1)
Y1=Y_train.reshape(-1,1)
reg.fit(X1,Y1)
```

```python
#for predicting the values -->
Y_predict = reg.predict(X_test)
print(Y_predict)
```

 #11. Visvalizing Training data -->

```python
plt.scatter(X_train,Y_train,color="red")
plt.plot(X_train,reg.predict(X_train),color="blue")
plt.title("")
plt.xlabel("")
plt.ylabel("")
plt.show()
```

 #12. Visvalizing testing data -->

```python
plt.scatter(X_test,Y_test,color="red")
plt.plot(X_train,reg.predict(X_train),color="blue")
plt.title(" ")
plt.xlabel(" ")
plt.ylabel(" ")
plt.show()
```

 #13. Model Evaluation -->

1.   Root Mean Squared Error.
2.   Explained variance score.
3.   R2 score.

```python
import sklearn.metrics as sm
print("Root mean squared error=",round(np.sqrt(sm.mean_squared_error(Y_test,Y_predict)),2))
print("Explained Variance Score=",round(sm.explained_variance_score(Y_test,Y_predict),2))
print("R2 Score=",round(sm.r2_score(Y_test,Y_predict),2))
```
