Breadth-first search is a graph traversal algorithm that starts traversing the graph from the root node and explores all the neighboring nodes. Then, it selects the nearest node and explores all the unexplored nodes. While using BFS for traversal, any node in the graph can be considered as the root node.

 #Method-1 --->

```python
#Tree-1

SuccList={'a':['b','c'],'b':['a','c','d'],'c':['a','b','d'],'d':['b','c']}     #keys == nodes , values == branches .
Start='a'
Goal='e'
Closed = list()       # contains nodes that are traversed.
SUCCESS=True
FAILURE=False
State=FAILURE
```

```python
def GOALTEST(N):
	if N == Goal:
		return True
	else:
		return False
```

```python
def MOVEGEN(N):               # called when we have move program pointer to another node.
	New_list=list()
	if N in SuccList.keys():
		New_list=SuccList[N]
	print("New_list=",New_list)
	return New_list
```

```python
def APPEND(L1,L2):
	New_list=L1+L2
	return New_list
```

```python
def BFS():                                   # Breadth-first search.
	OPEN=[Start]         #conatins nodes that we are going to traverse or we are at that node.
	CLOSED=list()        #contains nodes that are already traversed.
	global State
	global Closed
	while (len(OPEN) != 0) and (State != SUCCESS):          #stop if we are at last node or we got success
		print("------------")
		N= OPEN[0]
		print("N=",N)
		del OPEN[0] #delete the node we picked
		
		if GOALTEST(N)==True:
			State = SUCCESS
			CLOSED = APPEND(CLOSED,list(N))
			print("CLOSED=",CLOSED)
		else:
			CLOSED = APPEND(CLOSED,list(N))
			print("CLOSED=",CLOSED)
			CHILD = MOVEGEN(N)
			print("CHILD=",CHILD)
			for val in CLOSED:
				if val in CHILD:
					CHILD.remove(val)
			for val in OPEN:
				if val in CHILD:
					CHILD.remove(val)
			OPEN = APPEND(OPEN,CHILD) #append movegen elements to OPEN
			print("OPEN=",OPEN)
	Closed=CLOSED
	return State
```

```python
#Driver Code
result=BFS() #call search algorithm
print(Closed,result)
```

```python
# Tree-2

SuccList={'a':['b','c','d'],'b':['a','e','f'],'c':['g'],'d':['h','i'],'e':['j']}
Start='a'
Goal='j'

Closed = list()
SUCCESS=True
FAILURE=False
State=FAILURE

def GOALTEST(N):
	if N == Goal:
		return True
	else:
		return False

def MOVEGEN(N):
	New_list=list()
	if N in SuccList.keys():
		New_list=SuccList[N]
	print("New_list=",New_list)
	return New_list
	
def APPEND(L1,L2):
	New_list=L1+L2
	return New_list

def BFS():
	OPEN=[Start]
	CLOSED=list()
	global State
	global Closed
	while (len(OPEN) != 0) and (State != SUCCESS):
		print("------------")
		N= OPEN[0]
		print("N=",N)
		del OPEN[0] #delete the node we picked
		
		if GOALTEST(N)==True:
			State = SUCCESS
			CLOSED = APPEND(CLOSED,list(N))
			print("CLOSED=",CLOSED)
		else:
			CLOSED = APPEND(CLOSED,list(N))
			print("CLOSED=",CLOSED)
			CHILD = MOVEGEN(N)
			print("CHILD=",CHILD)
			for val in CLOSED:
				if val in CHILD:
					CHILD.remove(val)
			for val in OPEN:
				if val in CHILD:
					CHILD.remove(val)
			OPEN = APPEND(OPEN,CHILD) #append movegen elements to OPEN
			print("OPEN=",OPEN)
	Closed=CLOSED
	return State
	
#Driver Code
result=BFS() #call search algorithm
print(Closed,result)
```
