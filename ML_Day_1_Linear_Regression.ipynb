```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/House_price_predictionCopy.csv')

```

```python
from sklearn.linear_model import LinearRegression
#from sklearn import linear_model
#reg=linear_model.LinearRegression()
reg=LinearRegression()                   #linear regression is stored in a small variable'reg'.
x=df[['SQUARE_FT']]                                          
y=df['TARGET(PRICE_IN_LACS)']
reg.fit(x,y)                  #fit --> means training is done.
```

```python
plt.scatter(df['SQUARE_FT'],df['TARGET(PRICE_IN_LACS)'])   #scatter to mark dataset on graph.
plt.show()
print(reg.predict([[1275]]))            #square fit is passed inside predict to predict target price in lacs.
#predict is used for testing, Independent variable in the double square bracket {one as list , other as subscript}
```

```python
print("Coefficient",reg.coef_)                    #coefficient implies slope{m} in y=mx+c
print("Intercept",reg.intercept_)                 # intercept prints c.
```

```python
from sklearn.model_selection import train_test_split
X=df[['BHK_NO.','SQUARE_FT']]
y=df['TARGET(PRICE_IN_LACS)']
# test size 0.2 means 20 percent data will be taken as test data , and  train size is automatically taken as 0.8 ,i.e., 80% . 
# Apparently,only one out of test aur train data is specified, the other one is taken automatically.

#train_test_split ---> splits the given dataset into into test size like X has two values BHK and SQ.FIT ,both will be splited in 20% test data and 80% train data.
#Simiarily y will be splited too. 

# random_ state --> it specifies that jo data humne 20 aur 80% liya h,vo randmly pooore dataset me select kiya jayega.
#{Seed method in random or Random method generates random number which is unique every time.}

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
reg1=LinearRegression()
reg1.fit(X_train,y_train)
print(reg1.predict(X_test))

# Best fit line is the closest line , means is minimized.
```

```python
print(reg1.coef_)
print(reg1.intercept_)
```
