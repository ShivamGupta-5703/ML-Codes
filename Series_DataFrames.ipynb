```python
import pandas as pd
```

```python
print(pd.__version__)
```

```python
L=[10,20,30,40,50]
S=pd.Series(L)         # Seies is created with the help of keyword 'Series'.
print(S)
```

```python
print(S[3])    # we can print any entry from series.
```

```python
S=pd.Series(L,index=['a','b','c','d','e'])  #Changing numeric index to key index given by user.
print(S)
```

```python
S=pd.Series(L,index=[chr(i) for i in range(97,102)]) #passing key index by loops.
print(S)
```

```python
print(S['d'])          #printing any entry with key index.
```

```python
# printing series with dictionaries.
d={'A':11,'B':22,'C':33,'D':44,'E':55,'F':66}
S=pd.Series(d)
print(S)
```

```python
#Creating dataframes (2D).

d={'name':['shivam','rahul','madhav','aman'],'age':[10,20,30,40]}
df=pd.DataFrame(d)           #keyword 'DataFrame'.
print(df)                    #printing dataframe.

print('-------------------------------------------------------------')

#to extract a particular row from dataframe.
#print(df[2])      (gives error we cannot extract row from dataframe as series)
print(df.loc[2])       #we use 'loc' attribute to extract row from dataframe.
print('-------------------------------------------------------------')

S=pd.Series(d)               # for series 'name' and 'age' are acting as key index.
print(S)                     #printing series.

print('-------------------------------------------------------------')

print(S['age'])              # to extract row from given data. 
```

```python
# to extract multiple row with have to pass index in a list.
print(df.loc[[0,3]])
```

```python
# to extract multiple row we have to pass index in a list.
#using loops.
print(df.loc[[i for i in range(0,4)]])
print('-------------------------------------------------------------')
#using steps.
print(df.loc[[i for i in range(0,4,2)]])  
print('-------------------------------------------------------------')
# using slicing.   #upper boundary index is also included. 
print(df.loc[0:3])
print('-------------------------------------------------------------')
#Negative index willl work.
print(df.loc[-4:-1])
```

```python
#passing key index in dataFrames.
df1=pd.DataFrame(d,index=['a','b','c','d'])
print(df1)
```
