 #1. Importing libraries and loading dataset-->

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("/content/Salary_Data.csv")
df
```

 #2. Visvalize the dataset -->

```python
plt.scatter(df[["YearsExperience"]],df["Salary"],color='red')
plt.title("Linear Regression Salary vs experience")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.show()
```

 #3. Categorising dataset into independent and dependent variables -->

```python
#Variation 1 for initialising independent variable--> gives a 2-D array.
X=df.iloc[:,:-1].values
print(X)
print("-------------------------------------------------------------------")
print(X.ndim)
```

```python
#Variation 2 for initialising independent variable--> gives a 2-D array.
X=df.iloc[:,[0]].values
print(X)
print("-------------------------------------------------------------------")
print(X.ndim)
```

```python
#Variation 3 for initialising independent variable--> gives a 1-D array.
X=df.iloc[:,0].values
print(X)
print("-------------------------------------------------------------------")
print(X.ndim)
```

```python
#Variation 4 for initialising independent variable--> gives a 1-D array.
X=df[["YearsExperience"]].values
print(X)
print("-------------------------------------------------------------------")
print(X.ndim)
```

```python
#Initialising dependent variable--> gives a 1-D array.
Y=df.iloc[:,1].values
print(Y)
print("-------------------------------------------------------------------")
print(Y.ndim)
```

 #4. Training and Testing -->

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
print(X_train)
print("-------------------------------------------------------------------")
print(X_test)
print("-------------------------------------------------------------------")
print(Y_train)
print("-------------------------------------------------------------------")
print(Y_test)
```

 #5. Applying Linear Regression Model -->

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
X1=X_train.reshape(-1,1)
Y1=Y_train.reshape(-1,1)
reg.fit(X1,Y1)
```

```python
#for predicting the values -->
Y_predict = reg.predict(X_test)
print(Y_predict)
```

 #6. Analyzing the results -->

```python
Y_predict = Y_predict.flatten()
diff_pred=Y_test-Y_predict
res_df = pd.concat([pd.Series(Y_predict),pd.Series(Y_test),pd.Series(diff_pred)],axis=1)
res_df.columns = ["Predictions","Original Data","Difference"]
res_df
```

 #7. Visvalize the Training data -->




```python
plt.scatter(X_train,Y_train,color="red")
plt.plot(X_train,reg.predict(X_train),color="blue")
plt.title("Training of Linear Regression Salary vs Experience")
plt.xlabel("Years")
plt.ylabel("Salary")
plt.show()
```

 #8. Visvalize the Testing data -->

```python
plt.scatter(X_test,Y_test,color="red")
plt.plot(X_train,reg.predict(X_train),color="blue")
plt.title("Training of Linear Regression Salary vs Experience")
plt.xlabel("Years")
plt.ylabel("Salary")
plt.show()
```

 #9. Model Evaluation -->

1.   Root Mean Squared Error.
2.   Explained variance score.
3.   R2 score.



```python
import sklearn.metrics as sm
print("Root mean squared error=",round(np.sqrt(sm.mean_squared_error(Y_test,Y_predict)),2))
print("Explained Variance Score=",round(sm.explained_variance_score(Y_test,Y_predict),2))
print("R2 Score=",round(sm.r2_score(Y_test,Y_predict),2))
```

 #10. Salary Prediction -->

```python
new_salary_pred = reg.predict([[15]])
print(new_salary_pred)
```

Now, changing the training and testing size of dataset -->

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("/content/Salary_Data.csv")
df1

X=df1[["YearsExperience"]].values
print(X)
print("-------------------------------------------------------------------")

Y=df1.iloc[:,1].values
print(Y)
print("-------------------------------------------------------------------")

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
A=X_train.reshape(-1,1)
B=Y_train.reshape(-1,1)
reg.fit(A,B)

Y_predict = reg.predict(X_test)
print(Y_predict)
print("-------------------------------------------------------------------")

import sklearn.metrics as sm
print("Root mean squared error=",round(np.sqrt(sm.mean_squared_error(Y_test,Y_predict)),2))
print("-------------------------------------------------------------------")
print("Explained Variance Score=",round(sm.explained_variance_score(Y_test,Y_predict),2))
print("-------------------------------------------------------------------")
print("R2 Score=",round(sm.r2_score(Y_test,Y_predict),2))
print("-------------------------------------------------------------------")

new_salary_pred = reg.predict([[15]])
print(new_salary_pred)
```
