```python
from sklearn.datasets import load_digits
d=load_digits()
```

```python
import pandas as pd
df=pd.DataFrame(d['data'],columns=d['feature_names'])
```

```python
df['target']=d['target']
```

```python
X=df.drop('target',axis='columns')
y=df['target']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier 
model=RandomForestClassifier() 
model.fit(X_train,y_train)  
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
#classification_report --> tells 'precision','recall','f1-score',and 'support'.
#confusion_matrix --> used to print metrices.
#accuracy_score --> prints accuracy of model.

cm=confusion_matrix(y_test,y_pred)        #y_test --> tested data, y_pred --> predicted data.
print(cm)

# Diagonal data --> Actual = predicted.
# 1 shows error.
```

```python
# To make metrices attractive -->
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
import seaborn as sns
sns.heatmap(cm,annot=True)       # heatmap is used for giving a makeover to metrices. 'annot{annotations }= true'--> values should be filled in following metrices.
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

```python
print(classification_report(y_test,y_pred))           
```

```python
print(accuracy_score(y_test,y_pred))
```
