```python
#Here, we are using GridSearchCV.
#GridSearchCV tries all the combinations of values passed in dictionary and evaluates the model for each combination using "Cross Validation" method.
#"Cross Validation"--> is a technique in which we train our model using the subset of dataset and then evaluate using the complementary dataset.
#The best parameter values are extracted and then predictions are made.
```

```python
import pandas as pd
df=pd.read_csv('/content/Iris.csv')
```

```python
X=df.drop('Species',axis='columns')
y=df['Species']
```

```python
from sklearn.tree import DecisionTreeClassifier 
model=DecisionTreeClassifier()
```

```python
# Passing parameters in a form of dictionary. 
parameters={'criterion':['gini','entropy','log_loss'],          
            'splitter':['best','random'],
            'max_depth':[1,2,3,4,5],
            'max_features':['auto','sqrt','log2']}
```

```python
from sklearn.model_selection import GridSearchCV
new_model=GridSearchCV(model,param_grid=parameters,scoring='accuracy',cv=5)      #model, parameters and cv is passed in grid search cv.
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=9)
new_model.fit(X_train,y_train)
```

```python
print(new_model.best_params_)             #new_model.best_params_ --> selects best parameter among those passed in dictionary.   
```

```python
print(new_model.score(X_test,y_test))
```
