```python
#Importing K-Means -->
from sklearn.cluster import KMeans
import pandas as pd
df=pd.read_csv('income.csv')   
df
```

```python
#Plotting clusters of points from dataset using matplotlib -->
import matplotlib.pyplot as plt
plt.scatter(df['age'],df['income'])
plt.xlabel('age')           #x-axis label.
plt.ylabel('income')        #y-axis label.
plt.show()

#Here,3 clusters are formed. But it is not accurate. Some point of differnet clusters are positioned with other clusters.
#That's why we are using K-means algorithm.
```

```python
model=KMeans(n_clusters=3)                #Let k=3.{Appropriate of k is calculated by ELBOW Method.}       {GridSearchCV can also be used for calculating value of K.}
y_pred=model.fit_predict(df[['age','income']])               #Fit and predict can be done together by using "fit_predict"
print(y_pred)                     #since we are using 3 clusters ,thus lables are 0,1, and 2 for each clusters.
```

 #In Unsupervised Learing we do not have a "dependent variable".Thus, here we are adding a dependent variable named "cluster". 

```python
df['cluster']=y_pred
df
```

```python
# To find the coordinates (x,y) of Center of Gravity/Cluster Head /Centriod of the clusters we use "cluster_centers_".
#Here, e+01 means {*10^01}.
print(model.cluster_centers_)
```

```python
#Creating dataframe containing values of individual clusters from main dataframe.

df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]
```

```python
#Assigning different colors to different clusters -->

plt.scatter(df1['age'],df1['income'],color='red')
plt.scatter(df2['age'],df2['income'],color='green')
plt.scatter(df3['age'],df3['income'],color='blue')
plt.xlabel('age')
plt.ylabel('income')
plt.legend()
plt.show()

#Here, we can cleraly see,it is not accurate. Some point of differnet clusters are positioned with different clusters.
```

 #But the cluster is correct. It is visually incorrect becoz of the large differences in scales of X axis and Y axis. 


 #To change the scales we will perform preprocessing.

```python
# Here, we will try to bring sacle of X and Y between 0 and 1.

from sklearn.preprocessing import MinMaxScaler
dfnew=df                              #making copy of original dataset.
scaler=MinMaxScaler()                 #Assigning minmaxscaler constructor to scaler.

#Y-axis.
scaler.fit(dfnew[['income']])         #Fiting of dataset.
dfnew['income']=scaler.transform(dfnew[['income']])      #Transformation of scales.
#X-axis.
scaler.fit(dfnew[['age']])
dfnew['age']=scaler.transform(dfnew[['age']])

dfnew
```

```python
#Again using K-Means and plotting graph -->
model1=KMeans(n_clusters=3)
y_pred1=model1.fit_predict(dfnew[['age','income']])
print(y_pred1)
```

```python
#Center head will also get changed.
dfnew['cluster']=y_pred1
print(model1.cluster_centers_)
```

```python
#Creating dataframe containing values of individual clusters from main dataframe.
df11=dfnew[dfnew['cluster']==0]
df22=dfnew[dfnew['cluster']==1]
df33=dfnew[dfnew['cluster']==2]

#Assigning different colors to different clusters -->
plt.scatter(df11['age'],df11['income'],color='red')
plt.scatter(df22['age'],df22['income'],color='green')
plt.scatter(df33['age'],df33['income'],color='blue')

#This is used to print the center head of the clusters.
#[:,0] means all values of x and y in column X{index 0 for X}.
#[:,1] means all values of x and y in column Y{index 1 for Y}.
plt.scatter(model1.cluster_centers_[:,0],model1.cluster_centers_[:,1],marker='*',color='magenta')

plt.xlabel('age')
plt.ylabel('income')
plt.legend()
plt.show()
```

# ELBOW **METHOD**
To calculate the correct value k ,i.e., number of clusters.

```python
 # Yaha par humne ek khali list li, aur usme sum of squared error for each value of k store karwaya, 
 #k ki value 'i' assign kari aur usme 1 to 10 k ki values ki loop lagai.Usme training kari ,aur khali list me append kara diya.
 #Now, sse has sum of squared error for each value of k.
sse=[]
for i in range(1,11):
  model2=KMeans(n_clusters=i)
  model2.fit(df[['age','income']])
  sse.append(model2.inertia_)
```

```python
sse        #contains 10 value for each value for k.
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(range(1,11),sse)
plt.show()

#Now, jaha par bhi elbow turn aayega{sharp corner}, that is the best value of k.{in this case 3 is the best value of this model.}
```
