```python
import pandas as pd
df=pd.read_csv('/content/Iris.csv')


from sklearn.model_selection import train_test_split
X=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=df['Species']

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=10)

from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
y_predicted = model.predict(X_test)
print(y_predicted)
```

```python
model.score(X_test,y_test)         #this tells by accuracy.If we multiply by 100 it gives accuracy in percentage.
#Here, accuracy is 100% becoz dataset taken is too small and is trained well.
```

```python
from sklearn.datasets import load_iris     #importing dataset 
df=load_iris(as_frame=False)      # Returns a dictionary
# By default dictionary aati h.
# 'as_frame=False' or default ,i.e., 'load_iris()' prints data in dictionary format, if we want data in dataset format then 'as_frame=True'
print(df)

#in output data is key of dictionary
#'data' refers to X_train(training data). 'data' contains value of independent variables. 
#Iris datset had 4 independent values sepal length,width and petal length,width , thus each array in 'data' contains 4 values.
```

```python
print(list(df.keys()))            #To print all keys of dictionary in list format.
#If we are not printing keys in list format, then 'dict_keys' will be displayed before keys. 
```

```python
print(df['frame'])                     #To print values of any specific key.
# 'data' key will print all independent values in dictionary form.
# 'target' key will print values in form of 0,1,2. It categories versicolor ,setosa,virginica in 0,1and 2.
# 't'arget_names' will print names of categories ,i.e., versicolor ,setosa and virginica.
# 'frame' means dataframe. If we print 'frame' then it gives'none' , becoz we initialised 'df' as 'as_frame=False'. 
# 'DESCR' means description. It prints 'Dataset Characteristics' like full statistics,classification,SD, mean,min,max,missing attribute values.
# 'feature_names' are name of independent variables.
# 'filename' prints name of file.
# 'data_module' prints name of module {sklearn.datasets.data}.
```

```python
print(df['data'].shape)              #Printing shape of particular key.     
#We had 5 columns sepal length,width , petal length,width and species. But here,we are getting only 4 columns becoz data prints only independent values,and species is dependent variable.
```

```python
#Slicing. df[data] means independent values {X_train,X_test} and slicing [r1:r2,c1:c2].
#[:]--> means all records , [3:]--> means third column se saare column.

X=df['data'][:,3:]                                            #Difference between [3:] {isme 2D data aayega}and [3].reshape(-1,1) {isme 1 D data ayega}.
print(X)                                                       # Independent variables are always 2D.
print(X.shape)    #only values of single column are printed.
```

```python
y=df['target']      #{dependent variable(always 1D data)}.
print(y)
print(y.shape)
```

```python
y=df['target']==2                #Binary Classification{if 2 --> then true,else false}.
print(y)
print(y.shape)
```

```python
import numpy as np
y=(df['target']==2).astype(np.int)           #To convert true/false into integer 1/0 we use {astype},using module numpy. 
# 2 index mtlb virginica{target me virginca ko 2 assign kiya tha}.
print(y)
print(y.shape)
```

```python
from sklearn.linear_model import LogisticRegression 
model=LogisticRegression()               #Applying LOgistic Regression.
model.fit(X,y)                           #Trianing.
#X is last column printed above{independent variable}{petal width}, y bhi humne upar nikala h{dependent variable}.
tt=model.predict([[1.5]])                #Testing.    #Here 1.5 is petalwidth, we are giving petalwidth kyuki humne upar ussi ko print karwaya tha.
#Petal width is independent, so we have to give in 2D format, thus applying double square brackets.
print(tt)
# If answer is 1,then virginica and if 0,then not virginica.
```

```python
import numpy as np
#linspace takes (starting value, ending value,number of parts the value is to be divided{jitne points humne starting aur ending value ke beech me chhaiye.})
#linspace hume 1D data dega but X {petal width} independent value h, toh hum reshape karenge 1D array data ko 2D array me.
#We can convert upper dimentional data into lower dimension data and vice-versa,provided that no.of elements in source and destination must be equal with the help of reshape function.
# reshape(row,column) , here if dont know rows or column we specify it as '-1'{known as unknown dimension}.  reshape(-1,1) means unknown row and 1 column.

#Along x axis we have petal width.                   
xnew=np.linspace(0,3,1000).reshape(-1,1)       
#Along y axis we have probability{either 1 or 0}.
y_prob=model.predict_proba(xnew)             #it is a predefined function
#agar hume prediction karni h,then use 'predict' and for calculating probability use 'preduct_proba'.
print(xnew) 
print('--------------------')
print(y_prob)
```

```python
import matplotlib.pyplot as plt        #To plot line on graph.
plt.plot(xnew,y_prob[:,1])             #Taking all rows and column of index 1.

#We are getting squashed curve {sigmoid curve}.
#Here, we are getting perfect curve becoz we are using trained data.
```
