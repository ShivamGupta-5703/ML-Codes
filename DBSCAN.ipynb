```python
import pandas as pd
df=pd.read_csv('Iris.csv')
df
```

```python
X=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] #Independent variabkes.
```

```python
from sklearn.cluster import DBSCAN
model=DBSCAN(eps=0.7,min_samples=2)        #Hyperparameters Episilon {by default it is 0.5} and minimum points.
```

```python
model1=model.fit(X)          #Training.
```

```python
#Predict is used for making predictions for classification or regression.But here,we do not expect any value for classification or regression.
#Here we need to make clusters. So,prediction not required.
```

```python
#Now training is done{using fit},means clusters are also formed. 
#So to find number of "clusters" or "labels" we use --> "labels_ ".
labels=model1.labels_          
print(labels)

#Gives inofmration about each data record. NOT ACCURATE.
len(labels)  #gives 150.
```

```python
#Using numpy trying to find labels in True/False -->
import numpy as np
arr=np.zeros_like(labels,dtype=bool)        # Method "zeros_like" is used to create array of any dimension.
#While using "zeros" we have to provide dimension of required array.
#"zeros_like" copies the shape of old array. Like in this labels array used is a 1D array. So, new array formed will be 1D.
#Elements in zeros array craeted will all be "0".{Imp}
#"dtype=bool" means array contains "false".
arr
```

```python
#Here, we get all core points using pre-defined function "core_sample_indices_" and making their index "true".
arr[model1.core_sample_indices_]=True    
```

```python
arr
```

```python
#noc --> number of clusters.
#set removes duplicate values.
# So, labels has 150 records.
# "set(labels)" has 4 records {-1,0,1,2}
#-1 always represent noise point.
# "-(1 if -1 in labels else 0)" means agar labels me kahi bhi "-1" aata h ,then RHS=1 ,else RHS=0.

noc=len(set(labels))-(1 if -1 in labels else 0)

#Here, we removed our outlier{-1} from the dataset.
#Remaining values form clusters{i.e.,0,1,2}.
```

```python
print(noc)    #Lables had one "-1" has {4-1=3}.
```

```python
#Here we are using "silhouette_score" (comes under metrix) to find accuracy. Its value ranges between {-1  to  +1}.
#If result is positive and close to 1 ,it is good. If negative then bad, and -1 is worst. 0 for overlapping clusters.
from sklearn.metrics import silhouette_score
print(silhouette_score(X,labels))
```
