 #Method -1

```python
numQ=8;
currentSol = [0 for x in range(numQ)]
solution =[]
def isSafe(testrow, testcol):
  if testrow == 0:
   return True
  for row in range(0, testrow):
    if testcol == currentSol[row]:
      return False
    if (abs(testcol - currentSol[row]) == abs(testrow- row)):
      return False
  return True

for Q0 in range (numQ):
  if not isSafe (0,Q0):
    continue
  else :
    currentSol[0]=Q0 
  for Q1 in range (numQ):
    if not isSafe (1,Q1):
      continue
    else :
      currentSol[1]=Q1
    for Q2 in range (numQ):
      if not isSafe (2,Q2):
        continue
      else :
        currentSol[2]=Q2
      for Q3 in range (numQ):
        if not isSafe (3,Q3):
          continue
        else :
          currentSol[3]=Q3
        for Q4 in range (numQ):
          if not isSafe (4,Q4):
            continue
          else :
            currentSol[4]=Q4
          for Q5 in range (numQ):
            if not isSafe (5,Q5):
              continue
            else :
              currentSol[5]=Q5
            for Q6 in range (numQ):
              if not isSafe (6,Q6):
                continue
              else :
                currentSol[6]=Q6
              for Q7 in range (numQ):
                if not isSafe (7,Q7):
                  continue
                else :
                  currentSol[7]=Q7
                  solution.append(currentSol.copy())
```

```python
def PlaceQ(row):
  global currentSol,solution,numQ
  for col in range(numQ):
    if not isSafe(row,col):
      continue
    else:
       currentSol[row] == col
       if row== (numQ-1):
         solution.append(currentSol.copy())
       else:
           PlaceQ(row+1)

PlaceQ(0);
currentSol
len(solution)
```
