```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Iriscpy.csv')         #iriscpy has missing entries.
print(df)
```

```python
print(df.head(20))
```

```python
#'dropna' is used to remove rows with missing fields.
newdf=df.dropna()      #in this changes arre applied in new dataframe.   #agar hum dropped row excess karemge to erroe aayega.
print(newdf)                      #only 146 rows are printed.
```

```python
#if we want to apply changes in original file itself,then we use attribute(inplace==True).
df.dropna(inplace=True)               #(dropna=drop not available)
print(df.to_string())   
```

```python
#If we want to fill data in empty cell instead of dropping row, we use-> fillna(value that is to be filled)
df = pd.read_csv('/content/Iriscpy.csv')
df.fillna(5703,inplace=True)
print(df.head(15))
```

```python
#We can also give mean, median,mode inside of a particular value in fillna.
df=pd.read_csv('/content/Iriscpy.csv')
x=df['SepalWidthCm'].mean()                         #similarily, we can use median().
print("Mean is:",x)
df.fillna(x,inplace=True)
print(df.head(15))
#in this every empty cell is filled with mean becoz column is not mentioned with df while printing it.
```

```python
#to fill element in desired column we need to mention column name with df.
df=pd.read_csv('/content/Iriscpy.csv')
x=df['SepalWidthCm'].mean()                       
print("Mean is:",x)
df['SepalWidthCm'].fillna(x,inplace=True)
print(df.head(15))
```

```python
#for Mode to work we use subscript 0([0]) at th end for changes,without it no changes are applied to dataset.
df=pd.read_csv('/content/Iriscpy.csv')
x=df['SepalWidthCm'].mode()[0]                       
print("Mode is:",x)
df['SepalWidthCm'].fillna(x,inplace=True)
print(df.head(15))
```

```python
#if we want to delete empty cells in a particular column only , then we use -> attribute named 'subset=[column name]' inside dropna.
df=pd.read_csv('/content/Iriscpy.csv')
df.dropna(subset=['SepalWidthCm'],inplace=True)        #empty cells in only spealwidth are dropped.
print(df.head(15))
```

```python
#IN small data ,to modify any entry we use loc.
df=pd.read_csv('/content/Iriscpy.csv')

#df.loc[row,column]=value

df.loc[0,'SepalWidthCm']=999
df.head(15)     #entry in row 0 and column sepalwidth has changed.
```

```python
#But for large data we cannot change data manually,thus we use the concept of loops {we know the general threshold lower and upper bound.}
#Thus, we replace any abnormal entry with a constant entry.
df=pd.read_csv('/content/Iriscpy.csv')
for i in df.index:
  if df.loc[i,'SepalWidthCm']>3:
    df.loc[i,'SepalWidthCm']=55      #replacing value greater than 3 by 55 in sepalwidth.
df.head(15)
```

```python
#to remove Duplicate data, we use 'drop_duplicates'.
df=pd.read_csv('/content/Iriscpy.csv')
df.drop_duplicates(inplace=True)       #removes all duplicate data.
df.head()
```
