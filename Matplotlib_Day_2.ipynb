# **MARKERS**

'o'	Circle	
'*'	Star	
'.'	Point	
','	Pixel	
'x'	X	
'X'	X (filled)	
'+'	Plus	
'P'	Plus (filled)	
's'	Square	
'D'	Diamond	
'd'	Diamond (thin)	
'p'	Pentagon	
'H'	Hexagon	
'h'	Hexagon	
'v'	Triangle Down	
'^'	Triangle Up	
'<'	Triangle Left	
'>'	Triangle Right	
'1'	Tri Down	
'2'	Tri Up	
'3'	Tri Left	
'4'	Tri Right	
'|'	Vline	
'_'	Hline

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y,marker='p')
plt.show()
```

# **marker|line|color**

# **line**

'-'	Solid line	
':'	Dotted line	
'--'	Dashed line	
'-.'	Dashed/dotted line

# **color**

'r'	Red	
'g'	Green	
'b'	Blue	
'c'	Cyan	
'm'	Magenta	
'y'	Yellow	
'k'	Black	
'w'	White

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y,'D--r')      # markers|line-type|color.
plt.show()
```

other keyword parameters:-

## 1. **markersize** **ms**
## 2. **markeredgecolor** **mec**
## 3. **markerfacecolor** **mfc**

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y,'P--g',ms=25,markeredgecolor='k',markerfacecolor='r')
plt.show()
```

Alternative of **line|color** of **marker|line|color**

Attributes:-

**1. linestyle or ls**

**2. color or c**

Other Attribute is:

**1. linewidth or lw**


```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y,ls='dotted',c='c',linewidth=5)
plt.show()
```

# Plotting Multiples Lines with:-
**1. Single plot() Method**

**2. Multiple plot() method**

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
p=[10,20,30]
q=[5,10,15]
a=(3,4)
b=(9,16)
plt.plot(x,y,p,q,a)       
plt.show()
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
p=[10,20,30]
q=[5,10,15]
a=(3,4)
b=(9,16)
plt.plot(x,y)
plt.plot(p,q)
plt.plot(a,b)
plt.show()
```

# Attibutes of xlabel(), ylabel(), title():-
**1. fontdict={'family':'serif','color':'red','size':2}** for all 3 methods

**2. loc='left' or 'right' or 'center'**    for title() only

```python
x=[5,2,3,4,1]
y=[10,7,8,9,6]
plt.plot(x,y)

#to fing slope-->
#slope=(y[1]-y[0])/(x[1]-x[0])

# we can also loc to locate title,if not specified,it is center by default.
plt.title('GRAPH',fontdict={'family':'serif','color':'red','size':30})
#plt.title('GRAPH',fontdict={'family':'serif','color':'red','size':30},loc ='right')               #family- font type. 
plt.xlabel('X axis',fontdict={'family':'fantasy','color':'green','size':20})
plt.ylabel('Y axis',fontdict={'family':'DejaVu Sans','color':'magenta','size':20})
plt.show()
```

# Grid
Attributes:-

**1.axis='x' or 'y' or 'both' or None**

**2.color or c**

**3.linestyle or ls**

**4.linewidth or lw**

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y,c='g')
plt.grid(c='r',ls='dashed',lw=1)
plt.show()
```

# Subplot
**1.subplot(rows,cols,index)**

**2.suptitle()**

```python

plt.suptitle('Country')           #suptitle gives one title to whole document.
#plt.subplot(2,1,1)               #subplot used to define arrangment of graphs    subplot(row,column,index)
plt.subplot(1,2,1)                #subplot(121) is also correct.{order must be correct.}
plt.title('USA')                  #row tells number of row. same for column .index tells about which graph to assign underdefined entries or current position.
#plt.subplot(2,1,2)  
plt.subplot(1,2,2)              
plt.title('India')
plt.show()
```

```python

plt.suptitle('Country') 
plt.subplot(2,1,1)   
plt.title('USA')                  
plt.subplot(2,1,2)               
plt.title('India')
plt.show()
```

```python
plt.suptitle('Country') 
plt.subplot(121)   
plt.plot([1,2,3,4],[1,4,9,16],'D--r')             #this will mark points  along with specified line type in graph with assigned color.
plt.title('USA')                  
plt.subplot(122)      
plt.plot([1,2,3,4],[1,4,9,16],color='m')          #this will mark points in graph with assigned color,we also assign marker type.
plt.title('India')
plt.show()
```
