9. Finding mean, median and mode and replacing empty cells with them for individual column -->

```python
x = df["Maxpulse"].mean()        
y = df["Calories"].median()
z = df["Pulse"].mode()[0]
print("Mean of values in column 'Maxpulse'=",x)
print("Median of values in column 'Calories'=",y)
print("Mode of values in column 'Pulse'=",z)
df["Maxpulse"].fillna(x,inplace=True) 
df["Calories"].fillna(y,inplace=True)
df["Pulse"].fillna(z,inplace=True)
df.head(10)
```

10.   Converting all cells in the columns into the same format -->

```python
df["DOB"] = pd.to_datetime(df["DOB"])
print(df.head(10))
```

11. Removing empty values (NAT,null values)-->

```python
df["DOB"] = pd.to_datetime(df["DOB"])
df.dropna(subset=['DOB'],inplace=True)
print(df.head(10))
```

12. Fixing wrong data -->

```python
#(i) Replacing values -
for x in df.index:
  if df.loc[x,"Pulse"]>60:
    df.loc[x,"Pulse"]=50
df
```

```python
#(ii) Removing rows - 
for x in df.index:
  if df.loc[x,"Pulse"]>60:
    df.drop(x,inplace=True)
df
```
