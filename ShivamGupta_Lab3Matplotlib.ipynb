1. Importing Modules-->

```python
from matplotlib import pyplot as plt
from numpy import array as ar
import numpy as np
```

2. Plotting a graph -->

The plot() function is used to draw points (markers) in a diagram.

The keyword marker is used to draw each point with a specified marker.

```python
x = ar([1,2,3,4,5,6,7])
y = ar([5,6,7,8,9,0,4])
plt.plot(x,y,marker = "*")
plt.grid()
plt.xlabel("X-axis")                      # for naming x-axis.
plt.ylabel("Y-axis")                      # for naming y-axis.
plt.title("Graph")                        # for giving title of the graph.
plt.show()
```

If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3

```python
ypoints = ar([1,2,3,4,5,6,3])
plt.plot(ypoints,linestyle="dashed",color ="b")
plt.show()
```

3. Subplots -->

The subplot() function allows us to draw multiple plots.

```python
#plot 1 -->
x = ar([1,2,3,4,5,6])
y = ar([1,2,3,4,5,6])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.grid()
#plot 2 -->
x1 = ar([1,2,3,4,5,6])
y1 = ar([2,3,4,5,6,7])
plt.subplot(1,2,2)
plt.plot(x1,y1)
plt.grid()
plt.show
```

Variation of subplot -->

```python
x = ar([0, 1, 2, 3])
y = ar([3, 8, 1, 10])

plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.grid()

x = ar([0, 1, 2, 3])
y = ar([10, 20, 30, 40])

plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.grid()

x = ar([0, 1, 2, 3])
y = ar([3, 8, 1, 10])

plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.grid()

x = ar([0, 1, 2, 3])
y = ar([10, 20, 30, 40])

plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.grid()
plt.suptitle("hello")
plt.show()
```

```python
a = ar([1,2,3,4,5,6,7])
b = ar([1,3,5,7,9,11,13])

plt.suptitle('Subplot Graph')
plt.subplot(2, 3, 1)
plt.plot(a,b,color='red',marker='s',linestyle='dotted')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plot Line")

plt.subplot(2, 3, 2)
plt.plot(a,b)
plt.grid()
plt.title("Grid graph")

plt.subplot(2, 3, 3)
plt.scatter(a,b,marker='*',c='blue')
plt.title("Scatter Plot")

plt.subplot(2,3,4)
plt.title("Empty Graph")


plt.subplot(2,3,5)
plt.hist(a,width=0.1,color='blue')
plt.title("Histogram")

plt.subplot(2,3,6)
plt.bar(a,b,width=0.1,color='cyan')
plt.title("Bar Chart")
plt.show()

```

4. Scatter Plot -->

The scatter() function plots one dot for each observation.

```python
x = ar([1,2,3,4,5,6,7])
y = ar([5,6,7,8,9,0,4])
plt.scatter(x,y)
plt.show()
```

5. Bar Graph -->  

bar() function is used to draw bar graphs.

```python
x = ar([1,2,3,4,5,6,7])
y = ar([5,6,7,8,9,3,4])
plt.bar(x,y,color="red",width=0.7)
plt.show()
```

6. Histogram -->

hist() is used to make histograms.

A histogram is a graph showing frequency distributions.
It is a graph showing the number of observations within each given interval.

```python
x = np.random.normal(150,5,200)
plt.hist(x,width=2.8)
plt.show()
```

7. Piechart -->

The pie() function is used to draw pie charts.

```python
y = ar([23,27,34,16])
mylables = ["1","2","3","4"]
plt.pie(y,labels=mylables)
plt.show()
```

If Legends are not equal-->

```python
x = ar([12,23,23,18,9,15])
c = ar(['red','blue','yellow','green','cyan','hotpink'])
plt.pie(x,labels=[1,2,3,4,5,6],colors=c) 
plt.legend([1,2,3,4,5,6])
```

If labels are not equal-->

```python
x = ar([12,23,23,18,9,15])
c = ar(['red','blue','yellow','green','cyan','hotpink'])
plt.pie(x,labels=[1,2,3,4,5],colors=c) 
plt.legend([1,2,3,4,5,6])
```

If X and y are not equal -->

```python
x = ar([1,3,4,5,7,9])
y = ar([1,3,8,6,3])

plt.suptitle('Graph')
plt.subplot(2,2,1)
plt.plot(x,y,color='red',marker='o',linestyle='dashed')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plot Graph")
plt.show()
```

IF Linestyle is not defined(empty)-->

```python
x = ar([1,3,4,5,7])
y = ar([1,3,8,6,3])

plt.suptitle('Graph')
plt.subplot(2,2,1)
plt.plot(x,y,color='red',marker='o',linestyle='')
plt.show()
```
